```python
#encoding categorical variable (Male=0, Fenale=1)

df['Gender']= df['Gender'].map({'Male':0, 'Female':1})


#Selecting features for clustering

X=df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]


#Normalizing data

scaler= StandardScaler()
X_scaled= scaler.fit_transform(X)


# Determine the optimal number of clusters using the Elbow Method

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)


#Ploting the Elbow Method graph

plt.plot(range(1, 11), wcss, 'r-')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#From the plot, let's assume the optimal number of clusters is 5

optimal_clusters = 5


#Applying KMeans with the optimal number of clusters and explicitly set n_init

kmeans = KMeans(n_clusters=optimal_clusters, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X_scaled)


# Analyze and visualize the clusters

sns.pairplot(df, hue='Cluster', vars=['Age', 'Annual Income (k$)', 'Spending Score (1-100)'])
plt.show()
```

```python
# Summary statistics for each cluster

df.groupby('Cluster').mean()
```

```python
#Visualize the clusters in a 3D plot

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_scaled[:, 0], X_scaled[:, 1], X_scaled[:, 2], c=df['Cluster'], cmap='viridis')
ax.set_title('Customer Segments')
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income (k$)')
ax.set_zlabel('Spending Score (1-100)')
plt.colorbar(scatter)
plt.show()
```
